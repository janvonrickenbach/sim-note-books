# General funcs

```python
def get_var(sim,var):
   return max(0,sim.sim_info[var])
  
```

```python
def get_inf_conv(sim):
    gas.TPX = [sim.sim_info["temperature"],1E5,[0.1,0.0,0,0.897,0.003]]
    coeff = gas.mix_diff_coeffs[-1]
    return 1-np.exp(-coeff*sim.sim_info["sh_number"]/sim.sim_info["velocity"]*sim.sim_info["a_v"]*0.004)
```

```python
def get_grid(ratio=1.1,wc_thickness=1E-4,nx=32):
    dx_c = wc_thickness*(1-ratio)/(1-ratio**nx)
    dx = []
    for x_c in range(nx):
        dx.append(dx_c)
        dx_c *= ratio
    return np.cumsum(dx) -0.5*np.array(dx)
```

```python
def get_conv(sim):
    gas.TPX = [sim.sim_info["temperature"],1.015E5,[0.1,0,0,0.9-sim.sim_info["mf"]*1E-6,sim.sim_info["mf"]*1E-6]]
    mf_in=gas.Y[4]
    eval_factor = sim.sim_info["sh_number"]/2000. 
    eval_factor = 1
    eval_loc = 0.004
    try:
        avg_Y= np.interp(eval_loc/eval_factor,sim.wc_data["coords"],sim.wc_data["CO"])
        mf   =1-avg_Y/mf_in
    except:
       mf = -1

    return mf
```

```python
def create_simulation_python(input_dict,folder,root_folder=".",ref_folder=None):
     

    config_sim = jc.configuration_mb_local
    config_sim["transat_version"] = "wc_python"

    config_sim["jobname"] = inputs["folder"]
    config_sim["root_folder"] = root_folder
    simulation = sim_brutus.job(**config_sim)

    simulation.sim_info = input_dict.copy()
    simulation.sim_info["converged"] = False


    return simulation
```

```python
def load_wc(sim,L_r=0.005):
    wc_data = []
        
    time_idx = int(sim.sim_info["temperature"])

    nx = sim.sim_info["x_cells"]
    dx = L_r/ nx
    for point in range(sim.sim_info["x_cells"]):
      wc_data.append([point*dx+dx/2,np.genfromtxt(os.path.join(sim.info.local["sim_folder"],"result","grid_"+str(point)+"_"+str(time_idx)+"_"+str(sim.sim_info["cov_CO"])+".dat"))])
    
    sim.wc_data = {}
    sim.wc_data["coords"] = np.linspace(0.0,L_r,100)
    sim.wc_data["CO"]     = np.interp(sim.wc_data["coords"],[item[0] for item in wc_data],[np.array(item[1][4][0]) for item in wc_data])
    sim.all_data = wc_data  
```

```python
temp_arr = temperatures_ext


prod = []

prod = np.zeros([len(temp_arr),interface.n_reactions+1])
ref_prod = np.zeros([len(temp_arr),interface.n_reactions+1])


colors = ["r","r","b","b","k","k","m","m"]
linestyles = ["-","--","-","--","-","--","-","--"]

for reaction in range(interface.n_reactions+1):
    for idx,temp in enumerate(temp_arr[:]):
        sim =  filter_simulations(sim_list_py,temperature=temp,cov_CO=1,wc_geo_area=1,velocity=5,wc_thickness=-1E6)[0]
        conc = sim.all_data[0][1][4][1]
        gas.TPX = (temp,1.015E5,[0.1,0.064,0.072,0.76,conc])

        interface.TPX = (temp,1E5,[1,0,0,0])
        for reac in range(interface.n_reactions):
            interface.set_multiplier(1,reac)
        interface.advance_coverages_test(1000000)
        prod_ref = interface.net_rates_of_progress[4]
        if reaction == interface.n_reactions:
            interface.TPX = (temp+0.01*temp,1.0E5,[1,0,0,0])
            gas.TPX = (temp+0.00*temp,1.015E5,[0.1,0.064,0.072,0.76,conc])

        else:   
            interface.set_multiplier(1.01,reaction)
        interface.advance_coverages_test(1000000)

        prod[idx,reaction] = (interface.net_rates_of_progress[4]-prod_ref)/prod_ref/0.01
        conc_mol = gas.density * gas.Y[4]/0.028
```
