```python
def create_simulation_python(input_dict,folder,root_folder=".",ref_folder=None):
     

    config_sim = jc.configuration_mb_local
    config_sim["transat_version"] = "wc_python"

    config_sim["jobname"] = folder
    config_sim["root_folder"] = root_folder
    simulation = sim_brutus.job(**config_sim)

    simulation.sim_info = input_dict.copy()
    simulation.sim_info["converged"] = False


    return simulation
```

```python
def load_wc(sim):
    
    sim.all_data = {}
    for datatype in ["grid","fluxes"]:
        mf = sim.sim_info["mf"] 
        filename = os.path.join(sim.info.local["sim_folder"],"result",datatype+"_0_"
                                +str(int(sim.sim_info["temperature"]))+"_"+str(int(mf))+".dat")
        wc_data = np.genfromtxt(filename)
        sim.all_data[datatype] = wc_data  


def get_data(sim,time=False,flux=False):
    idx_t = 1
    if time:
        idx_t = 0
    key = "grid"
    if flux:
        key = "fluxes"
    return np.array([item for item in sim.all_data[key]])
```

```python
def plot_temp(temperatures,nx,name):
    fig, ax = pl.subplots()
    mf_real = massfractions/1E6
    data_list = [mf_real]
    for temp in temperatures[:]:
        label = str(temp) + " K"
        sim_to_plot = filter_simulations(sim_list_flux,temperature=temp,nx=nx)
        data = np.array([get_data(sim,flux=True)[4,0] for sim in sort_sims(sim_to_plot,"mf")])/0.028
        data_list.append(data)

        ax.plot(mf_real,data,label=label)


    ax.set_xlabel("CO Massfraction (-)")
    ax.set_ylabel("Reaction rate (mol/($m^2 s$)")
    ax.set_ylim([0,ax.get_ylim()[1]*1.5])
     #   pl.plot(mf_real,f(mf_real),'o')
    pl.legend(loc=0,ncol=2,frameon=False)
    pl.gcf().tight_layout()
    pl.gcf().savefig("reacrates.png",dpi=600)
    return data_list

data_list = []
for nx in grids:
    data_list.append(plot_temp(temperatures[:],nx,"reacrates"))


#plot_temp([500],512,"reacrates")

```
