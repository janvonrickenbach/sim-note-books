```python
def filter_simulations(sim_list,**kwargs):
  filtered_list = []
  for sim in sim_list:
      match_all = True
      for key, item in kwargs.iteritems():
            if not int(sim.sim_info[key]) == int(item):  
              match_all = False
      if match_all:
        filtered_list.append(sim)
  return filtered_list
    
def sort_sims(sim_list,prop): 
    return sorted(sim_list,key=lambda it: it.sim_info[prop])    

def get_array(sim_list,func=None):
    prop_list = []
    if func:
        for sim in sim_list:
           prop_list.append(func(sim))
    return np.array(prop_list)
```

```python
def create_simulation(input_dict,folder,root_folder="."):

    try:
        sh.copytree(ref_folder,os.path.join(root_folder,folder))
    except:
        print("directory exists")
 
    massfraction = {}
    massfraction["O2"]   = [0.1,   [0,0]] 
    massfraction["CO2"]  = [0.064, [0,1]]
    massfraction["H2O"]  = [0.072, [0,2]]
    massfraction["N2"]   = [0.76,  [0,3]]
    massfraction["CO"]   = [input_dict["mf"]*1E-6, [0,4]]
    velocity     = input_dict["velocity"]
    temperature  = input_dict["temperature"]
    wc_thickness = input_dict["wc_thickness"]*1E-6
    wc_geo_area  = input_dict["wc_geo_area"]
    sh_n         = input_dict["sh_number"]
    a_v          = input_dict["a_v"]
    n_wc         = input_dict["wc_ncells"]
    res          = input_dict["resolution"]
    dp           = input_dict["dp"]*1E-9
    cov_CO       = input_dict["cov_CO"]
    config_sim = jc.configuration_brutus
    config_sim["jobname"] = folder
    config_sim["root_folder"] = root_folder
    config_sim["cluster_scratch"]='/cluster/scratch_xp/public/vjan/wc_1d_tests'
    simulation = sim_brutus.job(**config_sim)
    

    sh.copy(os.path.join(os.path.join(root_folder,ref_folder),"transat_mb.inp"),os.path.join(root_folder,folder))
    sh.copy(os.path.join(os.path.join(root_folder,ref_folder),"finite_vol_avg.stt"),os.path.join(root_folder,folder))

    sh.copy(os.path.join(os.path.join(root_folder,ref_folder),"cantera_input.cti"),os.path.join(root_folder,folder))
    sh.copy(os.path.join(os.path.join(root_folder,ref_folder),"mt_coefficients.dat"),os.path.join(root_folder,folder))
    sh.copy(os.path.join(os.path.join(root_folder,ref_folder),"res_coefficients.dat"),os.path.join(root_folder,folder))

    simulation.application.change_input("cornerpoint_number  21",res,newline="cornerpoint_number "+str(res) +"\n",filename="finite_vol_avg.stt") 
    
    transatui_command = []
    transatui_command.append("transatui.py")
    transatui_command.append("--Project=finite_vol_avg.stt")
    transatui_command.append("--Graphics=bash")
#    ui=sp.Popen(transatui_command,cwd=os.path.join(root_folder,folder),stdout=open("ui_output.out","a"))
    ui=sp.Popen(transatui_command,cwd=os.path.join(root_folder,folder))

    ui.communicate() 
    
    Dp = 0.002
    ds = 0.0003
    porosity = 0.903
    #a_v = 10.33/(Dp+ds) * np.sqrt(1-porosity) -5.8/(Dp+ds)*(1.0-porosity)
  
    write_inital_conditions(os.path.join(root_folder,folder),massfraction,velocity,temperature)
    
    component_string =  "if (x[ii] < 0.0001){component_molefraction_0_4[ii] ="+str(input_dict["mf"]*1E-6)+";}else{component_molefraction_0_4[ii] =0.000;}"
    if cov_CO < 1:
        simulation.application.change_input("component_molefraction_0_4[ii]",res,newline=component_string + os.linesep,filename="initialconditions.cxx")        

    simulation.application.change_input("data_format",'\"paraview_cell\"')
    simulation.application.change_input("wc_thickness",str(wc_thickness))
    simulation.application.change_input("activate_chem_heat_src",'.false.')
    simulation.application.change_input("mf_subiter",'1')
    simulation.application.change_input("L3Dplot",'.true.')
    simulation.application.change_input("L2Dplot",'.false.')
    simulation.application.change_input("write_kplane",'.false.')
    simulation.application.change_input("ode_rtol",'1E-7')
    simulation.application.change_input("ode_atol",'1E-25')
    simulation.application.change_input("wc_geo_area",str(wc_geo_area))
    simulation.application.change_input("sherwood_number",str(sh_n))
    simulation.application.change_input("den(1)",0.028*1E5/(8.314*temperature))
    simulation.application.change_input("wc_ncells",str(n_wc))
    simulation.application.change_input("wc_pore_diameter",str(dp))
    simulation.application.change_input("wc_tortuosity",str(8.0))
    simulation.application.change_input("pseudo_int_time",str(1E4))
    simulation.application.change_input("init_surf_cov(1)",str(1-cov_CO))
    simulation.application.change_input("init_surf_cov(3)",str(cov_CO))
    simulation.application.change_input("applyeiinitial",".false.")
    
    if wc_thickness < 0.0:
        simulation.application.change_input("wc_ncells",str(1))
        simulation.application.change_input("wc_thickness",str(1E-5))

    simulation.application.change_input("porous_a_v",str(a_v))

    simulation.application.change_input("surface_coverage",'1.0',filename="properties.dat")

    conjugate = False
    if conjugate:
       simulation.application.change_input("temp_model",'\"conjugate\"',filename="properties.dat")
       simulation.application.change_input("density",'3500',filename="properties.dat")
       simulation.application.change_input("conductivity",'25.6',filename="properties.dat")
       simulation.application.change_input("heatcapacity",'795',filename="properties.dat")
       simulation.application.change_input("surface_coverage",'1.0',filename="properties.dat")


    simulation.sim_info = input_dict.copy()
    simulation.sim_info["converged"] = False


    return simulation
```

```python
def write_inital_conditions(folder,massfraction,velocity,temperature):

    writer = ini.initalconditions_writer()
    writer.add_variable("uvelocity",velocity)
    writer.add_variable("temperature",temperature)
    
    for key,item in massfraction.iteritems():
        writer.add_variable(var="component_molefraction",value=item[0],index=item[1])
        
    writer.write_file(folder)
```

```python
def get_folder_name(inputs):
  folder = ""
  for item in sorted(inputs.iteritems(), key=op.itemgetter(0)):
      folder += item[0][0:5]+"_"+str(int(item[1]))+"_"
  return folder[:-1]
        

sim_list_full = []
sim_list = []

input_dict = {} 
#temperatures = [360,400,440,450,460,470,480,490,500,510,520,530,540,550,560,570,600,700,800,900,1000,1100]
temperatures = [400,440,470,500,540,600,700,900,1100]
#temperatures = [560]

wc_ncells = [32]
res = [21]

par_inputs = {}


par_inputs["wc_thickness"] = [10,40,-1E6]
par_inputs["wc_geo_area"] = [100,1,10,500]
par_inputs["sh_number"] = [4000,2000]
par_inputs["velocity"] = [2.5,5]
par_inputs["a_v"] = [1000]
par_inputs["dp"] = [10,100]
par_inputs["mf"] = [3000,10000]
cov_co = [1,0]


root_folder = "/home/vonrickenbach/testcases_mb/kelvin_cell_final/wc_1d_tests/"

default_inputs = {}
for key,item in par_inputs.iteritems():
    default_inputs[key] = item[0] 

sim_count = 0
for temperature in temperatures:
    for resolution in res:
        for n_cells in wc_ncells:
          for co in cov_co:
            current_inputs = {}
            current_inputs = default_inputs.copy()
            current_inputs["resolution"] = resolution
            current_inputs["wc_ncells"] = n_cells
            current_inputs["temperature"] = temperature
            current_inputs["cov_CO"]      = co
            global_inputs = current_inputs.copy()
            sim_list_full.append(create_simulation(current_inputs,get_folder_name(current_inputs),root_folder=root_folder))
            sim_count += 1
            for par_input,par_values in par_inputs.iteritems():
                current_inputs = global_inputs.copy()
                for par_value in par_values[1:]:
                    current_inputs[par_input] = par_value
                    sim_count += 1
                    sim_list_full.append(create_simulation(current_inputs,get_folder_name(current_inputs),root_folder=root_folder))
    
sim_list = sim_list_full                    
```

```python
def filter_file(sim):
    sim.cp()
    filename = os.path.join(sim.info.local["sim_folder"],"std_out.log_cluster")
    file_obj = open(filename)
    last_job = []
    for line in file_obj:
        last_job.append(line)
        if "Your job looked like" in line:
            last_job = []
            
    for line in last_job:       
        if "---> iteration stops here" in line:
            sim.sim_info["converged"] = True
            return "converged"

    sim.sim_info["converged"] = False
    return None
    
```

```python
def get_conv(sim):
    try:
        avg_Y= np.interp(0.005,sim.wc_data["coords"],sim.wc_data["CO"])
        mf   = (0.00299-avg_Y)/0.00299
    except:
        mf = np.nan
    if mf <0.0:
        mf = np.nan
    return mf

```
