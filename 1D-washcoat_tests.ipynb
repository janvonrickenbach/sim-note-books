## General Functions

```python
def filter_simulations(sim_list,**kwargs):
  filtered_list = []
  for sim in sim_list:
      match_all = True
      for key, item in kwargs.iteritems():
            if not int(sim.sim_info[key]) == int(item):  
              match_all = False
      if match_all:
        filtered_list.append(sim)
  return filtered_list
    
def sort_sims(sim_list,prop): 
    return sorted(sim_list,key=lambda it: it.sim_info[prop])    

def get_array(sim_list,func=None,var=None):
    prop_list = []
    if func:
        for sim in sim_list:
           if var:
               prop_list.append(func(sim,var))
           else:
               prop_list.append(func(sim))
    return np.array(prop_list)
```

```python
def get_folder_name(inputs):
  folder = ""
  for item in sorted(inputs.iteritems(), key=op.itemgetter(0)):
      if item[0] in par_inputs.keys() or item[0] == "temperature" or item[0] == "cov_CO":
          folder += item[0][0:4]+"_"+str(int(item[1]))+"_"
  return folder[:-1]
```

```python
def filter_file(sim):
    sim.cp()
    filename = os.path.join(sim.info.local["sim_folder"],"std_out.log_cluster")
    file_obj = open(filename)
    last_job = []
    for line in file_obj:
        last_job.append(line)
        if "Your job looked like" in line:
            last_job = []
            
    for line in last_job:       
        if "---> iteration stops here" in line:
            sim.sim_info["converged"] = True
            return "converged"

    sim.sim_info["converged"] = False
    return None
    
```

```python
def get_conv(sim):
    gas.TPX = [sim.sim_info["temperature"],1.015E5,[0.1,0,0,0.9-sim.sim_info["mf"]*1E-6,sim.sim_info["mf"]*1E-6]]
    mf_in=gas.Y[4]
    eval_factor = sim.sim_info["sh_number"]/2000. 
    eval_factor = 1
    try:
        avg_Y= np.interp(0.004/eval_factor,sim.wc_data["coords"],sim.wc_data["CO"])

        mf   =1-avg_Y/mf_in
    except:
       mf = -1
#    if mf <0.0:
#        mf = np.nan
    return mf

```

```python
def get_var(sim,var):
   return max(0,sim.sim_info[var])
  
```

```python
def get_grid(ratio=1.1,wc_thickness=1E-4,nx=32):
    dx_c = wc_thickness*(1-ratio)/(1-ratio**nx)
    dx = []
    for x_c in range(nx):
        dx.append(dx_c)
        dx_c *= ratio
    return np.cumsum(dx) -0.5*np.array(dx)
```

```python
def write_inital_conditions(folder,massfraction,velocity,temperature):

    writer = ini.initalconditions_writer()
    writer.add_variable("uvelocity",velocity)
    writer.add_variable("temperature",temperature)
    
    for key,item in massfraction.iteritems():
        writer.add_variable(var="component_molefraction",value=item[0],index=item[1])
        
    writer.write_file(folder)
```

```python
def add_simulations(sim_list,sim_count,temperatures,wc_ncells,res,default_inputs,par_inputs,cov_co,add_ref=True):
  root_folder = "/home/vonrickenbach/testcases_mb/kelvin_cell_final/wc_1d_tests/"

  for temperature in temperatures:
    for resolution in res:
        for n_cells in wc_ncells:
          for co in cov_co:
            current_inputs = {}
            current_inputs = default_inputs.copy()
            current_inputs["resolution"] = resolution
            current_inputs["wc_ncells"] = n_cells
            current_inputs["temperature"] = temperature
            current_inputs["cov_CO"]      = co
            global_inputs = current_inputs.copy()
            print add_ref
            if add_ref:
                sim_list.append(create_simulation(current_inputs,get_folder_name(current_inputs),root_folder=root_folder))
                sim_count += 1
            for par_input,par_values in par_inputs.iteritems():
                current_inputs = global_inputs.copy()
                for par_value in par_values[1:]:
                    if par_input == "wc_thickness" and par_value == -1E6:
                        loc_wc_geo_area = par_inputs["wc_geo_area"]
                    else:
                        loc_wc_geo_area = [current_inputs["wc_geo_area"]]
                    for geo_a in loc_wc_geo_area:
                        
                        current_inputs["wc_geo_area"] = geo_a
                        current_inputs[par_input] = par_value
                        sim_count += 1
                        sim_list.append(create_simulation(current_inputs,get_folder_name(current_inputs),root_folder=root_folder))
    
```

```python
def get_sum(idx,prod,grid):
    return np.cumsum(prod[idx:]*np.diff(grid[idx:]))

def get_indices(idx,prod,grid):
    list_to_test = get_sum(idx,prod,grid)
    indices = np.where(list_to_test < 0.9*get_sum(0,prod,grid)[-1])[0]
    if len(indices) == len(list_to_test):
        return np.array([])
    else:
        return np.array([index + idx for index in indices])
    
def get_indices_list(prod,grid):
   return [get_indices(index,prod,grid) for index in range(len(prod))]

def sorted_list(prod,grid):
    dropped_empty = [item for item in get_indices_list(prod,grid) if len(item)> 0]
    return sorted(dropped_empty,key=lambda a:sum(np.diff(grid[a])) )
    
def get_first_nonzero(sorted_list):
    return [item for item in sorted_list][0]
   
```

```python
def load_wc(sim,L_r=0.005):
    wc_data = []
        
    time_idx = int(sim.sim_info["temperature"])
#    print sim.info.local["sim_folder"]

    nx = sim.sim_info["x_cells"]
    dx = L_r/ nx
    for point in range(sim.sim_info["x_cells"]):
      wc_data.append([point*dx+dx/2,np.genfromtxt(os.path.join(sim.info.local["sim_folder"],"result","grid_"+str(point)+"_"+str(time_idx)+"_"+str(sim.sim_info["cov_CO"])+".dat"))])
    
    sim.wc_data = {}
    sim.wc_data["coords"] = np.linspace(0.0,L_r,100)
    sim.wc_data["CO"]     = np.interp(sim.wc_data["coords"],[item[0] for item in wc_data],[np.array(item[1][4][0]) for item in wc_data])
    sim.all_data = wc_data  

```

```python
def get_time(name):
    name = os.path.basename(name)
    try:
        return int("".join([char for char in name[6:] if char.isdigit() or char == "-"]))
    except:
        print(name)
    
get_time("grid_0_-1.dat ")
```

```python
def load_wc_ani(sim):

    filelist = gl.glob(os.path.join(sim.info.local["sim_folder"],"grid_0_*"))
    
    
    filelist = [filename for filename in filelist if get_time(filename) >  0]
    print(os.path.join(sim.info.local["sim_folder"],"time_file.dat"))
    time_data = np.genfromtxt(os.path.join(sim.info.local["sim_folder"],"time_file.dat"))

    
    sorted_list = sorted(filelist,key=lambda name :get_time(name))
    wc_data = []
    for filename in sorted_list:
        wc_data.append([time_data[get_time(filename)-1,1],np.genfromtxt(filename)])
    
    sim.all_data = wc_data  
    return time_data
```

```python
def print_wc_time(test_sims,sp=4,sli=1,time=3.6E-5):

    nplots = 3
    figs = [pl.figure(figsize=(12,6)) for item in range(1)]
    axs = [figs[0].add_subplot(1,nplots,item+1) for item in range(nplots)]

    figs[0].tight_layout()
    ncells = 8
    
    for idx,test_sim in enumerate(test_sims[::-1]):
        data = np.array([item[1] for item in test_sim.all_data])
        time_val = np.array([item[0] for item in test_sim.all_data])

        
        axs[0].plot(time_val,data[:,-1,sli],'r')

        time_idx = min(np.where(time_val>time)[0])

        axs[0].axvline(time_val[time_idx])

        st_idx = 0
        grid = get_grid(nx=8)
        grid = np.insert(grid,0,0.0)
        axs[1].plot(grid[st_idx:],data[time_idx,-1,st_idx:])
        axs[2].plot(np.linspace(0,0.005,ncells+1)[st_idx:],data[time_idx,4,st_idx:])
        axs[0].set_title("T at first cell over time")
        axs[1].set_title("Temperature in wc")
        axs[2].set_title("CO in wc")
        axs[2].set_ylim([0,0.03])
```
