```python
class ListTable(list):
    """ Overridden list class which takes a 2-dimensional list of 
        the form [[1,2,3],[4,5,6]], and renders an HTML Table in 
        IPython Notebook. """
    
    def _repr_html_(self):
        html = ["<table>"]
        for row in self:
            html.append("<tr>")
            
            for col in row:
                html.append("<td>{0}</td>".format(col))
            
            html.append("</tr>")
        html.append("</table>")
        return ''.join(html)
```

```python
en_solid = 0
en_fluid = 1
nx = 40
L = 8.0
dx = L / nx
T_fluid_0 = 600.0
K = 1

K = 2.254*12*4.5E-5*1.5/(0.002)
D = 14.1*0.2/(0.002*1.5*1000*1.0)
```

```python
def get_T_fluid(mat,b,K,T_fluid_0,dx,y):
 
    b = -K*y
    b[0] = b[0] -T_fluid_0/dx
    
    return np.linalg.solve(mat,b)

def rhs(t,y,mat,b,K,T_fluid_0,dx,y_dot,D):
    T_f = get_T_fluid(mat,b,K,T_fluid_0,dx,y)

    y_dot[1:-1] = (y[:-2] - 2*y[1:-1] + y[2:])/dx**2 
    y_dot[0] = (-y[0] + y[1])/dx**2
    y_dot[-1] = (-y[-1] + y[-2])/dx**2

    return K*(T_f - y) + D*y_dot


            
        
```

```python
def integrate(K,D,stop_on_ig=True,time=20):
    mat = -(np.diag(np.ones(nx)) - np.diag(np.ones(nx-1),-1))/dx - np.diag(np.ones(nx)*K)
    b = np.ones(nx)
    y_dot = np.ones(nx)
    integ = si.integrate.ode(rhs)
    integ.set_f_params(*[mat,b,K,T_fluid_0,dx,y_dot,D])
    integ.set_initial_value(np.ones(nx)*300)
    dt=0.05
    while integ.successful() and integ.t < time:
        sol = integ.integrate(integ.t+dt)
        if sol[0] > 550 and stop_on_ig:
            break
    T_f = get_T_fluid(mat,b,K,T_fluid_0,dx,sol)
    return sol,T_f,integ.t
```

```python
d_arr = np.linspace(0,2.0,20)
k_arr = np.linspace(0.2,2.0,20)

dd,kk = np.meshgrid(d_arr,k_arr)

result = np.ones(dd.shape)
result_avg =result.copy()
for idx_d,d in enumerate(d_arr):
   for idx_k,k in enumerate(k_arr):
      result[idx_k,idx_d]= integrate(k,d)[2]
      
      result_avg[idx_k,idx_d]= np.sum(integrate(k,d)[0])

```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111)


p=ax.contourf(dd,kk,result,rstride=1,cstride=1,cmap=pl.cm.coolwarm,levels=np.linspace(0,np.max(result),21))

for temp,k,d,marker in zip([300,550],K,D,["o","x"]):
   for  case,name,color in zip(range(3),["low $L_{Kelvin}$","ref","$\epsilon$ = 0.9"],["r","k","b"]):
        ax.scatter(d[case],k[case],color=color,label=name,marker=marker)
        val = interp2d(d_arr,k_arr,result)(d[case],k[case])[0]

        ax.text(d[case],k[case],s="  t'="+str(val)[0:3])

ax.axis('tight')

cb = fig.colorbar(p)
cb.set_label("Time to ignition (-)")
ax.set_xlabel("$Pe_s$")
ax.set_ylabel("$N_{ht}$")
ax.set_xlim([0,1.4])

ax.legend(frameon=False,scatterpoints=1,loc=2,ncol=2)
fig.tight_layout()
fig.savefig("pics_3/T_max.pdf")
```
