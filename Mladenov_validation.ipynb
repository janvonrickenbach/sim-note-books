```python
import numpy as np
import matplotlib.pyplot as pl
import cantera as ca
import os
import scipy as sc
%matplotlib inline
```

```python
def get_rate(T,c):
    k0 = 1E12
    K0 = 4.5E5/1E6
    E  = 15000*4.184
    H  = 2000*4.184
    R = 8.134
    rr = k0*T*np.exp(-E/(R*T))*c*0.1*1E5/(R*T)/(1+K0*T*np.exp(H/(R*T))*c)**2
    return rr
```

```python
rates = []
for c in np.logspace(-4,-1,100):
    rates.append(get_rate(470,c))
```

```python
with_wc = True

if with_wc:
    nx = 160
    ref_folder = "/home/vonrickenbach/testcases_mb/kelvin_cell_final/wc_1d_tests/coupled_with_wc/"
else:
    nx = 40
    ref_folder = "/home/vonrickenbach/testcases_mb/kelvin_cell_final/wc_1d_tests/coupled_no_wc/"

gas = ca.import_phases(os.path.join(ref_folder,"cantera_input.cti"),["gas"])[0]
interface = ca.Interface(os.path.join(ref_folder,"cantera_input.cti"),"pt_surf",[gas])
```

```python


import glob as gl

filelist=gl.glob(os.path.join(ref_folder,"grid*"))
#filelist=gl.glob("/home/vonrickenbach/testcases_mb/kelvin_cell_final/wc_1d_tests/local_wc_test/grid*")

wc_data_l = []
for file_h in filelist:
    x_coord = 0.04/nx*int(os.path.basename(file_h).split("_")[1])
    save = int(os.path.basename(file_h).split("_")[2].split(".")[0])
    if x_coord > 0.04 or save < 0:
        continue
    wc_data_l.append([x_coord,np.genfromtxt(file_h)])

wc_data = [[item[0],item[1]] for item in sorted(wc_data_l,key=lambda item:item[0])]

wc_data_mol = np.zeros([len(wc_data),2,gas.n_species+interface.n_species,2])
for x_idx,wc_x in enumerate(wc_data):
    wc_data_mol[x_idx,0] = wc_x[0]
    for column in range(2):
        massfractions = wc_x[1][:gas.n_species,column]
        gas.Y = (massfractions)
        wc_data_mol[x_idx,1,:gas.n_species,column] =gas.X
        massfractions = wc_x[1][gas.n_species:interface.n_species+gas.n_species,column]
        interface.Y = (massfractions)
        wc_data_mol[x_idx,1,gas.n_species:interface.n_species+gas.n_species,column] =interface.X
```

```python
data_folder = "/home/vonrickenbach/papers/catalytic_convertor/validation_data/mladenov_2009/"

data = {}

if not with_wc:
    data["NO"] = np.genfromtxt(os.path.join(data_folder,"inst_diffusion_no.dat"))
    data["C3H6"] = np.genfromtxt(os.path.join(data_folder,"inst_diffusion_c3h6.dat"))
    data["NO2"] = np.genfromtxt(os.path.join(data_folder,"inst_diffusion_no2.dat"))

else:
    data["NO"] = np.genfromtxt(os.path.join(data_folder,"NO_diff.dat"))
    data["NO2"] = np.genfromtxt(os.path.join(data_folder,"NO2_diff.dat"))
    data["C3H6"] = np.genfromtxt(os.path.join(data_folder,"C3H6_diff.dat"))
```

```python
species_list = ["NO","C3H6","NO2"]
colors = ["k","b","r"]
fig = pl.figure()
ax  = fig.add_subplot(1,1,1)
fig1 = pl.figure()
ax1  = fig1.add_subplot(1,1,1)
for co,species in zip(colors,species_list):

    sp_idx = (gas.species_names+interface.species_names).index(species)
    dx = 0.04/nx
    x_coord = np.linspace(0,0.04-dx,nx)
    x_coord = x_coord + dx/2
#    pl.plot(x_coord,[item[1][sp_idx][1] for item in wc_data_mol],label="surface")
    sim_data = [item[1][sp_idx][0] for item in wc_data_mol]
    ax.plot(x_coord,sim_data,color=co,label=species)
    ax.plot(data[species][:,0],data[species][:,1],"o",color=co,markeredgecolor=co,label=species+" Mladenov et al.")
    interp_data = sc.interp(x_coord,data[species][:,0],data[species][:,1])
    ax1.plot(x_coord,(interp_data-sim_data)/interp_data,color=co)
    ax.set_xlabel("x (m)")
    ax.set_ylim(0,0.00007)
    ax.set_xlim(0,0.04)

    ax.set_ylabel("Molefraction (-)")
 #   ax.legend()
if with_wc:
    name = "Mladenov_wc.png"
else:
    name = "Mladenov_no_wc.png"
fig.tight_layout()
fig.savefig(name,dpi=300)
```

```python
wc_data_mol = np.zeros([len(grid_data),gas.nSpecies()+interface.nSpecies(),2])

for x_idx in range(len(grid_data)):
    for column in range(2):
        massfractions = grid_data[x_idx,:gas.nSpecies(),column]
        gas.setMassFractions(massfractions)
        wc_data_mol[x_idx,:gas.nSpecies(),column] =gas.moleFractions()
        massfractions = grid_data[x_idx,gas.nSpecies():interface.nSpecies()+gas.nSpecies(),column]
        interface.setMassFractions(massfractions)
        wc_data_mol[x_idx,gas.nSpecies():interface.nSpecies()+gas.nSpecies(),column] =interface.moleFractions()
```
