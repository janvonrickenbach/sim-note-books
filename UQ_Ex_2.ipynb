```python
import numpy as np
import matplotlib.pyplot as pl
import scipy.linalg as la
import scipy.stats as st

%matplotlib inline
```

```python
from matplotlib.patches import Ellipse

fig = pl.figure()
ax  = fig.add_subplot(1,1,1,aspect="equal")

obj = []
label = []
color = ["r","k","b","g"]
for rho, co in zip([0,0.1,0.5,0.9],color):
#for rho in [0]:

    C = np.array([[1.0, rho],[rho, 1.0]])
    lamb,eigv=la.eig(C)
    ell = Ellipse(xy=(3,3),width=1./lamb[0],height=1./lamb[1],angle=np.rad2deg(np.arccos(eigv[0, 0])-np.pi/2),color=co)

    ell.set_facecolor('none')
    ax.add_artist(ell)


    obj.append(ell)
    label.append(str(rho))
    
ax.legend(obj,label,loc=2)
ax.set_xlim(-1,7)
ax.set_ylim(-1,7)
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")

fig.savefig("cov.png")
```
