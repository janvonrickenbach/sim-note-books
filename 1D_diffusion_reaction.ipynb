```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as pl

%load_ext sympyprinting
%pylab inline
sy.init_printing()
```

```python
x,Y,C1,C2,C3,C4,A,B,large,K,D,ReSc = sy.var('x Y C1 C2 C3 C4 A B large K D ReSc')
Y = Y(x)
```

```python
Pe = sy.var("Pe",positive=True)
K = sy.var("K",positive=True)
D = sy.var("D",positive=True)
ReSc = sy.var("ReSc",positive=True)
```

```python
diff_eq=sy.Eq(Y.diff(x),(D)*Y.diff(x,x)-K*Y )
```

```python
diff_eq
```

```python
gen_solution= sy.dsolve(diff_eq,Y)
gen_solution
```

```python
cond1=gen_solution.subs({x:0}).rhs
cond2=C1

eq1 = sy.Eq(cond1,1)
eq2 = sy.Eq(cond2,0)


boundary_cond=sy.solve((eq1,eq2),C1,C2)
```

```python
sol=gen_solution.subs(boundary_cond)
```

```python
sol
```
